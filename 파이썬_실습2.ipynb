#넘파이 배열과 파이썬 리스트 비교

```python
# 파이썬 리스트 자료 구조 데이터 생성
list_data = [1,2,3]
print(f'파이썬 리스트 {list_data}')

print('-'*80)

#넘파이 배열 생성
array_data = np.array(list_data)
print(f'넘파이 배열로 변환한 후 데이터: {array_data}')
```

```python
# 파이썬 리스트와 넘파이 배열 차이점 확인

## 덧셈 연산

### 파이썬 리스트
result1 = list_data + list_data
print(f'파이썬 리스트에 대한 덧셈 연산의 결과: \n{result1}')

print('-'*80)

## 넘파이 배열
result2 = array_data + array_data
print(f'넘파이 배열에 대한 덧셈 연산의 결과: \n{result2}')
print('-'*80)
## 곱셈 연산

## 파이썬 리스트
result3 = list_data*3
print(f'파이썬 리스트에 대한 단순 곱셈 연산의 결과:{result3}')

## 함수를 이용한 파이썬 리스트 곱셈
result4 =[]
for num in list_data:
    result4.append(num*3)

print(f'파이썬 리스트에 대한 곱셈 연산의 결과: {result4}')

print('-'*80)

result5 = array_data*3
print(f'넘파이 배열에 대한 곱셈 연산의 결과: {result5}')

print('-'*80)

## 통계 처리 (평균값 구하기)

### 파이썬 리스트
list_average =sum(list_data) / len(list_data)
print(f'파이썬 리스트에 대한 평균값: {int(list_average)}')
print('-'*80)
### 넘파이 배열
array_average =array_data.mean()
print(f'넘파이 배열에 대한 평균 값 : {array_average}')
```

```python
# 대괄호가 중첩된 리스트 자료 구조 생성
# 조건: 성분 원소의 갯수가 동일
list_data = ([[1,2,3],[1,2,3],[2,3,4]])
print(f'대괄호가 중첩된 리스트 자료 구조 데이터 확인: \n{list_data}')

print('-'*80)

# 넘파이 배열로 변환 -> 엑셀과 같은 표(table: 행과 열) 구조 형성-> 엑셀과 같은 표(talbe: 행과 열) 구조 생성
list_data = np.array([[1,2,3],[1,2,3],[2,3,4]])
print(f'대괄호가 중첩된 리스트를 넘파이 배열로 배열한 결과: \n{list_data}')
```

```python
# 문제 14: for문과 if-elif-else
# ● scores 리스트 [88, 95, 72, 59, 100]를 for문으로 순회하며, 각 점수에 대해 90점 이상은 'A',
# 80점 이상은 'B', 70점 이상은 'C', 그 외는 'F' 등급을 출력하세요.
# 예시: scores = [88, 95, 72, 59, 100]

scores = [88,95,72,59,100]
for scores in scores:
    if scores >= 90:
        print('A')
    elif scores >= 80:
        print('B')
    elif scores >= 70:
        print('C')
    else:
        print('F')
```

```python
# 문제 15: 딕셔너리, for문, get()
# ● 상품 가격표 prices = {'사과': 1500, '바나나': 3000, '체리': 5000}와 구매 목록 wishlist =
# ['바나나', '사과', '딸기']가 있습니다. for문으로 wishlist를 순회하며, prices 딕셔너리에서
# .get()으로 가격을 찾아 출력하세요.
# 예시: prices = {'사과': 1500, '바나나': 3000, '체리': 5000}
# wishlist = ['바나나', '사과', '딸기']

prices = {'사과': 1500, '바나나': 3000, '체리': 5000}
wishlist = ['바나나', '사과', '딸기']

for item in wishlist:
    price = prices.get(item, '가격 정보 없음') # .get() 메서드에 키와 기본값 전달
    print(f'{item}의 가격: {price}')
```

```python
# 문제 22: 데이터 정제 및 새로운 리스트 생성
# ● raw_data = [' apple ', ' banana\n', ' \ncherry'] 리스트를 생성 후 각 과일 이름에 포함된
# 불필요한 공백과 줄바꿈 문자를 strip()으로 제거하고, 깨끗해진 과일 이름만
# clean_fruits라는 새 리스트에 append()하여 최종 결과를 출력하세요.
raw_data = [' apple ', ' banana\n', ' \ncherry']

clean_fruits = []
for item in raw_data:
    cleaned_item = item.strip().strip('\n')
     # strip()으로 공백 제거 후, 다시 strip('\n')으로 줄바꿈 문자 제거
    clean_fruits.append(cleaned_item)
print(clean_fruits)
```

```python
# 문제 24: 회원 등급 자동 부여

# ● users 리스트에는 회원들의 구매 횟수(visits) 정보가 담겨 있습니다. for문으로 각 회원을
# 확인하고, 구매 횟수가 10회 이상이면 grade를 'VIP'로, 5회 이상이면 'Gold'로, 그 외는
# 'Silver'로 users 딕셔너리에 새로운 키-값 쌍을 추가한 뒤, 전체 users 리스트를 출력하세요.
users = [
{'id': 'user1', 'visits': 12},
{'id': 'user2', 'visits': 3},
{'id': 'user3', 'visits': 7} ]

for user in users:
    if user['visits'] >= 10:
        user['grade'] = 'VIP'
    elif user['visits'] >= 5:
        user['grade'] = 'Gold'
    else:
        user['grade'] = 'Silver'
print(users)

```

```python
# 문제 27: 이메일 주소에서 아이디만 추출하기
# ● 이메일 주소 리스트 emails가 있습니다. 각 이메일 주소에서 @를 기준으로 split()하여
# 아이디 부분만 user_ids라는 새 리스트에 담아 출력하세요.

emails = ['python@google.com', 'java@oracle.com', 'swift@apple.com']
user_ids = []
for email in emails:
    # email.split('@')는 '@'를 기준으로 문자열을 나누어 리스트를 반환합니다.
    # 예를 들어 'python@google.com'에 대해 ['python', 'google.com']을 반환합니다.
    # [0]은 이 리스트의 첫 번째 요소, 즉 아이디 부분을 선택합니다.
    user_ids.append(email.split('@')[0])
print(user_ids)

# 이해를 돕기 위해 예시를 추가합니다:
# email_example = 'test@example.com'
# print(f"'{email_example}'.split('@') 결과: {email_example.split('@')}")
# print(f"첫 번째 요소 (아이디): {email_example.split('@')[0]}")
# email.split('@')`는 이메일 문자열을 `@` 앞부분과 뒤부분, 이렇게 두 부분으로 이루어진 리스트로 분리합니다.
# 예를 들어, '[redacted link]'의 경우 `email.split('@')`는 `['python', 'google.com']`을 반환합니다.
# 우리는 이 리스트의 첫 번째 요소인 사용자 이름을 얻으려고 하므로 `[0]`을 사용하여 접근합니다.
```

```python
# 문제 28: 문자열 데이터를 딕셔너리로 변환
# ● "name:김철수,age:30,city:서울" 형태의 문자열 user_data가 있습니다. 이 문자열을
# split()과 for문을 활용하여 {'name': '김철수', 'age': '30', 'city': '서울'} 형태의 딕셔너리로
# 만들어 출력하세요.
user_data = "name:김철수,age:30,city:서울"
dic_data = {}

# 먼저 콤마(,)를 기준으로 문자열을 분리하여 각 '키:값' 쌍을 얻습니다.
pairs = user_data.split(',')

# 각 쌍을 순회하며 콜론(:)을 기준으로 다시 분리하여 키와 값을 추출하고 딕셔너리에 추가합니다.
for pair in pairs:
    key, value = pair.split(':') # '키:값' 문자열을 콜론을 기준으로 분리
    dic_data[key] = value

print(dic_data)
```

```python
# 문제 29: 대소문자 섞인 데이터 정규화
# ● tags = [' Python', 'JAVA ', ' \nRuby', 'python'] 리스트에 정리되지 않은 태그들이 있습니다.
# for문으로 각 태그의 공백을 strip()으로 제거하고, lower()로 모두 소문자로 바꾼 뒤,
# clean_tags 리스트에 append()하여 clean_tags 전체를 출력하세요.
# 예시: tags = ['  Python', 'JAVA  ', ' \nRuby', 'python']

tags = ['  Python', 'JAVA  ', ' \nRuby', 'python']
clean_tags = []
for tag in tags:
    clean_tags.append(tag.strip().lower())
    print(clean_tags)
```
